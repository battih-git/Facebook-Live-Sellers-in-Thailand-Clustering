 ### Problem statement: Perform clustering analysis on the Facebook pages of 10 Thai fashion and cosmetics retail sellers to categorize and understand patterns in user engagement across various post types (videos, photos, statuses, and links). Utilize metrics such as comments, shares, and reactions to group these pages into distinct clusters based on similarities in user interaction behaviors. 

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Data/Live_20210128.csv') #Load the dataset
```

```python
df.head() # Print first 5 observation of the data
```

```python
df.info() 
# Therea re 7050 observations (post) details with 16 columns. There are 3 columns which are completly blank.
```

```python
# Removing the unwanted columns
df.drop(['status_id','status_published','Column1','Column2','Column3','Column4'],axis=1,inplace=True)
```

```python
df
```

```python
# Converting the categorical to numerical variable for feature status type
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['status_type'] = encoder.fit_transform(df['status_type'])
```

```python
df
```

```python
for col in df.iloc[:,1:].columns:
    sns.boxplot(data=df,x='status_type',y=col,palette='Set2')
    plt.title(f"Box distribution of {col} for Status type")
    plt.show()
```

```python
# Scaling the values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_col = df.drop('status_type',axis=1).columns
df_scaled = scaler.fit_transform(df[num_col])
df_scaled = pd.DataFrame(df_scaled,columns=num_col)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(df_scaled)
kmeans.inertia_
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,max_iter=1000)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("Number of clusters (K)")
plt.ylabel("Inertia")
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
cluster_labels = kmeans.fit_predict(df_scaled)
kmeans.inertia_
```

```python
df['Cluster'] = cluster_labels
```

```python
cluster_summary = df.groupby('Cluster')[num_col].mean()
print("\nCluster Summary:")
cluster_summary
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(df_scaled,cluster_labels)
print(f'Silhouette score is: {score}')
```

* Cluster 0
    * This cluster seems to have relatively lower average values across all features compared to the other clusters

* Cluster 1
    * This cluster exhibits significantly higher average values for all features compared to other clusters
    * It has exceptionally high values for number of reactions, shares and comments.
    
* Cluster 2
    * This cluster has moderate to high average values for number of reactions and relatively low average values for number of comments and shares
    
* Cluster 3
    * This cluster has higher average value for number of comments and relatively less number of reactions and shares.

* Cluster 0 can be classified as Low engaging posts.
* Cluster 1 can be classified as High engaging posts.
* Cluster 2 can be classified as Moderate engaging post with high number of reactions.
* Cluster 3 can be classified as Moderate engaging post with high number of comments
